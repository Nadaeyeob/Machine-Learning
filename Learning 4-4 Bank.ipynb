### Santander Bank Preprocessing

```python
import pandas as pd
bank = pd.read_csv('C:/Users/sw993/OneDrive/바탕 화면/ML/bank_train.csv')
```

```python
bank.drop(['ID'],inplace=True,axis=1)
bank['var3'].replace(-999999,2,inplace=True)
```

```python
bank.head(3)
```

```python
bank_feature = bank.iloc[:,:-1]
bank_target = bank.iloc[:,-1]
```

```python
# train_test_split할 경우 skfold가 어느정도 보정되는 건지 확인
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(bank_feature,bank_target,test_size=0.2)
print(y_train.value_counts()/y_train.count())
print(y_test.value_counts()/y_test.count())
```

```python
# StartifiedKFold 복습
from sklearn.model_selection import StratifiedKFold
skfold = StratifiedKFold(n_splits=2)
n=1
for train_index,test_index in skfold.split(bank_feature,bank_target):
    X_train_sk, y_train_sk = bank_feature.iloc[train_index],bank_target.iloc[train_index]
    X_test_sk, y_test_sk = bank_feature.iloc[test_index],bank_target.iloc[test_index]
    print(n,'번째')
    n+=1
    print(y_train_sk.value_counts()/y_train_sk.count())
    print(y_test_sk.value_counts()/y_train_sk.count())
```

### XGBoost (XGBClassifier)

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

xgb_clf = XGBClassifier(n_estimators = 1000, learning_rate = 0.6)
xgb_clf.fit(X_train,y_train, early_stopping_rounds=100, eval_metric='auc', eval_set=[(X_train,y_train),
                                                                                   (X_test,y_test)])
xgb_roc_score = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('{0:.4f}'.format(xgb_roc_score))
```

```python
from sklearn.model_selection import GridSearchCV

xgb_clf = XGBClassifier()
params = {'n_estimators': [100],
         'max_depth' : [5,7],
         'min_child_weight' : [1,3],
         'colsample_bytree' : [0.5,0.75]}
grid_cv = GridSearchCV(xgb_clf, param_grid = params)
grid_cv.fit(X_train,y_train, early_stopping_rounds=100, eval_metric='auc',
           eval_set=[(X_train,y_train),(X_test,y_test)])
print(grid_cv.best_params_)
xgb_roc_score = roc_auc_score(y_test,grid_cv.predict_proba(X_test)[:,1],average = 'macro')
print('{0:.4f}'.format(xgb_roc_score))
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

xgb_clf = XGBClassifier(n_estimatos=1000, colsample_bytree = 0.75, max_depth = 5,
                       min_child_weight = 1, reg_alpha=0.03, learning_rate=0.02)
xgb_clf.fit(X_train,y_train,early_stopping_rounds = 200, eval_set\
           = [(X_train,y_train),(X_test,y_test)], eval_metric='auc')
xgb_roc_score = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1],average='macro')
print('{0:.4f}'.format(xgb_roc_score))
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,8))
plot_importance(xgb_clf, ax=ax, max_num_features=20,height=0.4)
```

### LightGBM (LGBMClassifier)

```python
from lightgbm import LGBMClassifier
lgbm_clf = LGBMClassifier(n_estimator=500)
evals = [(X_test,y_test)]
lgbm_clf.fit(X_train,y_train,early_stopping_rounds=100,eval_metric='auc',eval_set=evals,verbose=True)
lgbm_roc_score = roc_auc_score(y_test,lgbm_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC {0:.4f}'.format(lgbm_roc_score))
```

```python
from sklearn.model_selection import GridSearchCV
lgbm_clf = LGBMClassifier(n_estimators=200)
params = {'num_leaves': [32, 64 ],
          'max_depth':[128, 160],
          'min_child_samples':[60, 100],
          'subsample':[0.8, 1]}
gridcv = GridSearchCV(lgbm_clf, param_grid=params)
gridcv.fit(X_train, y_train, early_stopping_rounds=30, eval_metric="auc",
           eval_set=[(X_train, y_train), (X_test, y_test)])
print('GridSearchCV 최적 파라미터:', gridcv.best_params_)
lgbm_roc_score = roc_auc_score(y_test, gridcv.predict_proba(X_test)[:,1], average='macro')
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```

```python
lgbm_clf = LGBMClassifier(n_estimators=1000, num_leaves=32, sumbsample=0.8, min_child_samples=100,
                          max_depth=128)

evals = [(X_test, y_test)]
lgbm_clf.fit(X_train, y_train, early_stopping_rounds=100, eval_metric="auc", eval_set=evals,
                verbose=True)

lgbm_roc_score = roc_auc_score(y_test, lgbm_clf.predict_proba(X_test)[:,1],average='macro')
print('ROC AUC: {0:.4f}'.format(lgbm_roc_score))
```

```python
from lightgbm import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig,ax = plt.subplots(figsize=(10,12))
plot_importance(lgbm_clf,ax=ax,max_num_features=20)
```
