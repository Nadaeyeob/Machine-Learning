#### app_train과 app_test를 합쳐서 한번에 데이터 preprocessing 수행. 

```python
# datatype이 object인 항목에 관해 label encoding 진행
apps = pd.concat([app_train,app_test],axis=0)
```

```python
apps.head(3)
```

```python
apps.tail(3)
```

#### Object feature들을 Label Encoding
* pandas의 factorize()를 이용

```python
# datatype = object 꺼낸 후, 각 col 별 factorize 진행
cols = apps.dtypes[apps.dtypes=='object'].index.tolist()
for col in cols:
    # factorize의 경우 2개의 열로 구성되어 있음. (변환 숫자, 숫자에 해당하는 원본 Data)
    apps[col] = pd.factorize(apps[col])[0]
    
print(apps.dtypes[apps.dtypes==object])
```

```python
apps.info()
```

#### Null값 일괄 변환

```python
apps.isnull().sum().head(100)
```

```python
# -999로 모든 컬럼들의 Null값 변환
apps = apps.fillna(-999)
```

```python
apps.isnull().sum().sum()
```

```python
apps.shape
```

#### 학습 데이터와 테스트 데이터 다시 분리

```python
# 위에서 Null 값을 -999로 변환했기 때문에 Target 값을 -999로 변환하여 분리
app_test = apps[apps['TARGET'] == -999]
app_train= apps[apps['TARGET'] != -999]
print(app_test.shape, app_train.shape)
```

```python
app_train_train = app_train.drop('TARGET',axis=1)
app_train_target = app_train['TARGET']
```

#### 학습 데이터를 검증 데이터로 분리하고 LGBM Classifier로 학습 수행. 
* 피처용 데이터와 타겟 데이터 분리
* 학습용/검증용 데이터 세트 분리

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(app_train_train,app_train_target,test_size=0.2)
```

```python
from lightgbm import LGBMClassifier

clf = LGBMClassifier(
        n_jobs=-1,
        n_estimators=1000,
        learning_rate=0.02,
        num_leaves=32,
        subsample=0.8,
        max_depth=12,
        silent=-1,
        verbose=-1
        )

clf.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_valid, y_valid)], 
eval_metric= 'auc', verbose= 100, early_stopping_rounds= 50)
```

##### Feature importance 시각화 

```python
from lightgbm import plot_importance

plot_importance(clf, figsize=(16, 32))
```
