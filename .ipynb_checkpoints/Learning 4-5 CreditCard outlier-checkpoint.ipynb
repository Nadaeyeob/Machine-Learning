### CreditCard Fraud

```python
import pandas as pd
fraud = pd.read_csv('C:/Users/sw991/perpect guide/4장/creditcard.csv')
fraud.head(3)
```

```python
print(fraud['Class'].value_counts())
print(fraud['Class'].value_counts()/fraud['Class'].count())
```

```python
# Time Drop, Amount Drop 필요, Amount의 경우 StandardScaler, log1p 사용하여 Scaler 작업
# Scaler 작업 이후 outlier 제거 실시

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
amount = fraud['Amount'].values.reshape(-1,1)
scaler.fit(amount)
amount_scaled_np = scaler.transform(amount)

import numpy as np
amount_scaled_np
amount_scaled = pd.DataFrame(amount_scaled_np, columns=['Amount_scaled'])
amount_scaled.head(3)
```

```python
fraud = pd.concat([fraud,amount_scaled],axis=1)
fraud.head(3)
```

```python
fraud_df = fraud.drop(['Time','Amount_scaled'],axis=1)
fraud_scaled_df = fraud.drop(['Time','Amount'],axis=1)
```

```python
fraud_df.head(3)
```

```python
fraud_scaled_df.head(3)
```

```python
# Scale 전 확인
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(fraud_df.drop(['Class'],axis=1),fraud_df['Class'],test_size=0.2)
# print(X_train.count(),y_train.count(),X_test.count(),y_test.count())
print(y_train.unique())
```

```python
from xgboost import XGBClassifier
xgb_clf = XGBClassifier(n_estimator=1000)
eval_set = [(X_test,y_test)]
xgb_clf.fit(X_train,y_train,eval_metric='auc',eval_set=eval_set,early_stopping_rounds=100)
from sklearn.metrics import roc_auc_score
roc_score = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1])
print(roc_score)
```

```python
# Scale 후 확인, Scale 전 ROC_AUC_Score : 96.11  / Scale 후 ROC_AUC_Score : 98.61
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(fraud_scaled_df.drop(['Class'],axis=1),fraud_scaled_df['Class'],test_size=0.2)
# print(X_train.count(),y_train.count(),X_test.count(),y_test.count())
from xgboost import XGBClassifier
xgb_clf = XGBClassifier(n_estimator=1000)
eval_set = [(X_test,y_test)]
xgb_clf.fit(X_train,y_train,eval_metric='auc',eval_set=eval_set,early_stopping_rounds=100)
from sklearn.metrics import roc_auc_score
roc_score = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1])
print(roc_score)
```

```python
def log(df=None):
    df_copy = df.copy()
    amount_n = np.log1p(df_copy['Amount'])
    df_copy.insert(0,'Amount_log',amount_n)
    df_copy.drop(['Time','Amount','Amount_scaled'],axis=1,inplace=True)
    return df_copy
```

```python
fraud_log_df = log(df=fraud)
```

```python
# Scale 전 ROC_AUC_Score : 96.11  / Scale 후 ROC_AUC_Score : 98.61 / Log Scale 후 ROC_AUC_Score : 97.31
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(fraud_log_df.drop(['Class'],axis=1),fraud_log_df['Class'],test_size=0.2)
# print(X_train.count(),y_train.count(),X_test.count(),y_test.count())
from xgboost import XGBClassifier
xgb_clf = XGBClassifier(n_estimator=1000)
eval_set = [(X_test,y_test)]
xgb_clf.fit(X_train,y_train,eval_metric='auc',eval_set=eval_set,early_stopping_rounds=100)
from sklearn.metrics import roc_auc_score
roc_score = roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1])
print(roc_score)
```

### OUTLIER 확인

```python
import numpy as np

def get_outlier(df=None, column=None, weight=1.5):
    # fraud에 해당하는 column 데이터만 추출, 1/4 분위와 3/4 분위 지점을 np.percentile로 구함. 
    fraud = df[df['Class']==1][column]
    quantile_25 = np.percentile(fraud.values, 25)
    quantile_75 = np.percentile(fraud.values, 75)
    
    # IQR을 구하고, IQR에 1.5를 곱하여 최대값과 최소값 지점 구함. 
    iqr = quantile_75 - quantile_25
    iqr_weight = iqr * weight
    lowest_val = quantile_25 - iqr_weight
    highest_val = quantile_75 + iqr_weight
    
    # 최대값 보다 크거나, 최소값 보다 작은 값을 아웃라이어로 설정하고 DataFrame index 반환. 
    outlier_index = fraud[(fraud < lowest_val) | (fraud > highest_val)].index
    
    return outlier_index
```
