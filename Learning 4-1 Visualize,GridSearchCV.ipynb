### Classifier 시각화

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
%matplotlib inline

X_feature, y_label = make_classification(n_features=2, n_redundant=0, n_informative=2,
                                         n_classes=3, n_clusters_per_class=1)
```

```python
print(X_feature.shape)
```

```python
plt.scatter(X_feature[:,0],X_feature[:,1],marker='o', c = y_label, s=25,
           cmap = 'rainbow', edgecolor='k')
```

```python
import numpy as np

def visualize_boundary(model,X,y):
    fig,ax = plt.subplots()
    ax.scatter(X[:,0],X[:,1], c=y, s=25, cmap='rainbow', edgecolor='k',
              clim=(y.min(),y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
    model.fit(X,y)
    # meshgrid -> 축을 나눠줌 / linspace(A,B,C) -> A,B 사이를 C만큼 나눔
    xx,yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),
                        np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(),yy.ravel()]).reshape(xx.shape)
    
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx,yy,Z,alpha=0.3, levels=np.arange(n_classes+1) - 0.5,
                          cmap='rainbow', clim=(y.min(),y.max()),
                          zorder=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

# 특정한 트리 생성 제약없는 결정 트리의 Decsion Boundary 시각화.
dt_clf = DecisionTreeClassifier().fit(X_feature, y_label)
visualize_boundary(dt_clf, X_feature, y_label)
```

### Titanic 복습 및 Feature_importance 확인

```python
titanic_df = pd.read_csv('C:/Users/sw993/PerfectGuide-master/1장/train.csv')
```

```python
titanic_df.head(3)
```

```python
titanic_df.info()
# Cabin, PassengerId, Name, Ticket 제거
```

```python
titanic_df.drop(['Cabin','PassengerId','Name','Ticket'],inplace=True, axis=1)
```

```python
# Embarked 는 get dummys로 encoder 처리, Fare는 Standardscaler 처리
# Sex 는 male = 0, female =1 처리, Age는 평균 처리
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Sex'] = titanic_df['Sex'].apply(lambda x : 0 if x == 'male' else 1)
```

```python
encoder = pd.get_dummies(titanic_df['Embarked'])
```

```python
encoder.info()
```

```python
titanic_df.info()
```

```python
titanic_df_1 = pd.concat([titanic_df,encoder],axis=1)
```

```python
titanic_df_1.drop(['Embarked'],inplace=True, axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
Fare = titanic_df['Fare'].values.reshape(-1,1)
scaler.fit(Fare)
Fare_scaled = scaler.transform(Fare)
```

```python
Fare_df = pd.DataFrame(data = Fare_scaled, columns=['Fare_scaled'])
```

```python
titanic_df_2 = pd.concat([titanic_df_1,Fare_df],axis=1)
```

```python
titanic_df_2.drop('Fare',inplace=True,axis=1)
```

```python
titanic_df_2.head(3)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
dt_clf = DecisionTreeClassifier()
params = {
    'max_depth' : [6,8,10,12],
    'min_samples_split' : [2,4,6,8]
}
X_train = titanic_df_2.iloc[:,1:]
y_train = titanic_df_2.iloc[:,0]
```

```python
titanic_df_2.head(3)
```

```python
X_train.head(3)
```

```python
y_train.head(3)
```

```python
grid_cv = GridSearchCV(dt_clf,param_grid = params, scoring = 'accuracy' ,cv=3, verbose=1)
grid_cv.fit(X_train,y_train)
print(grid_cv.best_score_)
print(grid_cv.best_params_)
```

```python
best_estimator = grid_cv.best_estimator_
```

```python
feature_importance = best_estimator.feature_importances_
feature_df = pd.Series(feature_importance, index = X_train.columns)
```

```python
feature_df.sort_values(ascending=False,inplace=True)
```

```python
import seaborn as sns
plt.figure(figsize=(8,6))
plt.title('Titanic Feature_Importance')
sns.barplot(x=feature_df, y=feature_df.index)
plt.show()
```
