### 교차검증 ###

```python
from sklearn.model_selection import KFold
import numpy as np

dt_clf = DecisionTreeClassifier()
kfold = KFold(n_splits=5)
cv_accuracy=[]
```

```python
# KFold의 경우 numpy 에서 실행됌, DataFrame으로 진행시 Error
n_iter = 0
feature = iris.data
label = iris.target
for train_index, test_index in kfold.split(feature):
    # KFold 객체에 split() 호출 시 Fold 별 Index를 List 형태로 반환
    X_train, X_test = feature[train_index], feature[test_index]
    y_train, y_test = label[train_index], label[test_index]
    
    dt_clf.fit(X_train,y_train)
    pred = dt_clf.predict(X_test)
    n_iter +=1
    
    accuracy = np.round(accuracy_score(y_test,pred),4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print('n_iter : {0}. accuracy : {1:.4f}, train : {2}, test : {3}'\
          .format(n_iter,accuracy,train_size,test_size))
    print('n_iter : {0}, test : {1}'.format(n_iter,test_index))
    
    cv_accuracy.append(accuracy)
print('Mean of accuray_score : {0:.4f}'.format(np.mean(cv_accuracy)))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
skfold = StratifiedKFold(n_splits=5)

dt_clf = DecisionTreeClassifier()

n_iter = 0
cv_accuracy = []
for train_index, test_index in skfold.split(feature,label):
    # KFold의 split과 달리 label 까지 넣어줘야함
    X_train, X_test = feature[train_index], feature[test_index]
    y_train, y_test = label[train_index], label[test_index]
    
    dt_clf.fit(X_train,y_train)
    pred = dt_clf.predict(X_test)
    
    # accuracy_score를 다시 호출해주지 않아서 오류 발생 -> 해당 Cell 내 re-import하여 해결
    accuracy = np.round(accuracy_score(y_test,pred),4)
    cv_accuracy.append(accuracy)
    n_iter +=1
    
    print('n_iter : {0}, accuracy_score : {1:.4f}'.format(n_iter,accuracy))
    print('test : ', test_index)

print('mean of accuracy score : {0:4f}'.format(np.mean(cv_accuracy)))
```

### GridSearchCV

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

# 데이터를 로딩하고 학습데이타와 테스트 데이터 분리
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, 
                                                    test_size=0.2, random_state=121)
dtree = DecisionTreeClassifier()

### parameter 들을 dictionary 형태로 설정
parameters = {'max_depth':[1, 2, 3], 'min_samples_split':[2,3]}
```

```python
import pandas as pd

# param_grid의 하이퍼 파라미터들을 3개의 train, test set fold 로 나누어서 테스트 수행 설정.  
### refit=True 가 default 임. True이면 가장 좋은 파라미터 설정으로 재 학습 시킴.  
grid_dtree = GridSearchCV(dtree, param_grid=parameters, cv=3, refit=True, 
                          return_train_score=True)

# 붓꽃 Train 데이터로 param_grid의 하이퍼 파라미터들을 순차적으로 학습/평가 .
grid_dtree.fit(X_train, y_train)

# GridSearchCV 결과는 cv_results_ 라는 딕셔너리로 저장됨. 이를 DataFrame으로 변환
scores_df = pd.DataFrame(grid_dtree.cv_results_)
scores_df[['params', 'mean_test_score', 'rank_test_score', 
           'split0_test_score', 'split1_test_score', 'split2_test_score']]
```

```python
scores_df
```

```python
print(grid_dtree.best_params_)
print('{0:.4f}'.format(grid_dtree.best_score_))
```

```python
dt_clf_best = grid_dtree.best_estimator_
pred = dt_clf_best.predict(X_test)
print(np.round(accuracy_score(y_test,pred),4))
```
